## 迪杰斯特拉算法求最短路径

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
# G.add_nodes_from([1,2,3,4,5,6])
G.add_weighted_edges_from([
    (1,6,100),
    (1,5,30),
    (1,3,10),
    (2,3,5),
    (3,4,50),
    (4,5,20),
    (4,6,10),
    (5,6,60)
])

pos = nx.spring_layout(G, iterations=20)
nx.draw_networkx_edge_labels(G,pos)
plt.show()
```
